**Claim 1: age affects death rate**


---



```r
#changing values of the death column into 0,1 only
df$death_d <- as.integer(df$death != 0)
# calculating the death rate
sum(df$death_d) / nrow(df)
```

```r
alive=subset(df,death_d == 1)
dead=subset(df,death_d == 0)
mean(alive$age, na.rm = TRUE)
mean(dead$age,na.rm = TRUE)
```

```r
#Testing the significance of the result
t.test(alive$age, dead$age, alternative="two.sided", conf.level = 0.99)
```

        Claim was true

        The p-value is approximately 0, which means that the observed difference between the groups is so extreme that it is extremely unlikely to have occurred by random chance alone.

**Claim 2: Gender has no effect**


---



```r
men = subset(df, gender == "male")
women = subset(df, gender == "female")
A=mean(men$death_d, na.rm = TRUE)
B=mean(women$death_d, na.rm = TRUE)
cat("death rate of men is: ", A*100,"%" )
cat("
")
cat("death rate of women is: ", B*100,"%" )
```

```r
#Testing the significance of the result
t.test(men$death_d, women$death_d, alternative="two.sided", conf.level = 0.99)
```

     Claim was wrong

      The p-value is approximately 0, there is a genuine difference in the average percentage of deaths between men and women indicating that gender affects rate of deaths.
